```python
import ast
import pickle
import pandas as pd
import numpy as np
import re
from datetime import datetime
```

```python
# Calculate performance metrics given results file (LLM)
d_test = 'test1'
model = 'mistral-7b-finetuned'
k = 1
file_name_results = d_test + '_' + model + '_' + str(k) + '-shot' + '.pickle'
file_name_true = 'data_' + d_test + '.csv'

# Load prediction results
with open(file_name_results, 'rb') as file:
    information_extractions = pickle.load(file)
    
supp_name_pred = []
supp_vat_pred = []
date_pred = []
amount_pred = []
for extraction in information_extractions:
    try:
        supp_name_pred.append(ast.literal_eval(extraction)['Supplier company name'])
        supp_vat_pred.append(ast.literal_eval(extraction)['Supplier VAT number'])
        date_pred.append(ast.literal_eval(extraction)['Date'])
        try:
            amount_pred.append(float(ast.literal_eval(extraction)['Amount']))
        except:
            amount_pred.append(float(ast.literal_eval(extraction)['Amount'].replace(',','.'))) 
    except:
        try:
            extraction = extraction.strip()
            supp_name_pred.append(ast.literal_eval(extraction)['Supplier company name'])
            supp_vat_pred.append(ast.literal_eval(extraction)['Supplier VAT number'])
            date_pred.append(ast.literal_eval(extraction)['Date'])
            try:
                amount_pred.append(float(ast.literal_eval(extraction)['Amount']))
            except:
                amount_pred.append(float(ast.literal_eval(extraction)['Amount'].replace(',','.'))) 
        except:
            try:
                extraction = re.findall('{\n.*\n.*\n.*\n.*\n.*}', extraction)[0].strip()
                supp_name_pred.append(ast.literal_eval(extraction)['Supplier company name'])
                supp_vat_pred.append(ast.literal_eval(extraction)['Supplier VAT number'])
                date_pred.append(ast.literal_eval(extraction)['Date'])
                try:
                    amount_pred.append(float(ast.literal_eval(extraction)['Amount']))
                except:
                    amount_pred.append(float(ast.literal_eval(extraction)['Amount'].replace(',','.'))) 
            except: 
                supp_name_pred.append('')
                supp_vat_pred.append('')
                date_pred.append('')
                amount_pred.append('')
        
data_dict = {
    'supp_name_pred': supp_name_pred,
    'supp_vat_pred': supp_vat_pred,
    'date_pred': date_pred,
    'amount_pred': amount_pred
    }
df_pred = pd.DataFrame(data=data_dict)

# Load the true data set
df_true = pd.read_csv(file_name_true)
```

```python
# Supplier name accuracy
supp_name_acc = np.sum(df_pred['supp_name_pred'] == df_true['supp_name'])/df_true.shape[0]

# Error analysis
print('Accuracy')
print(supp_name_acc)
col = 'supp_name'
col_pred = 'supp_name_pred'
true_res = df_true[df_pred[col_pred] != df_true[col]][col]
pred_res = df_pred[df_pred[col_pred] != df_true[col]][col_pred]
df_error_analysis = pd.concat([true_res, pred_res], axis=1)
print(df_error_analysis)
```

```python
# Supplier VAT accuracy
supp_vat_acc = np.sum(df_pred['supp_vat_pred'] == df_true['supp_vat'])/df_true.shape[0]

# Error analysis
print('Accuracy')
print(supp_vat_acc)
col = 'supp_vat'
col_pred = 'supp_vat_pred'
true_res = df_true[df_pred[col_pred] != df_true[col]][col]
pred_res = df_pred[df_pred[col_pred] != df_true[col]][col_pred]
df_error_analysis = pd.concat([true_res, pred_res], axis=1)
print(df_error_analysis)
```

```python
# Date accuracy
date_acc = np.sum(df_pred['date_pred'] == df_true['date'])/df_true.shape[0]

# Error analysis
print('Accuracy')
print(date_acc)
col = 'date'
col_pred = 'date_pred'
true_res = df_true[df_pred[col_pred] != df_true[col]][col]
pred_res = df_pred[df_pred[col_pred] != df_true[col]][col_pred]
df_error_analysis = pd.concat([true_res, pred_res], axis=1)
print(df_error_analysis)
```

```python
# Amount accuracy
amount_acc = np.sum(df_pred['amount_pred'] == df_true['amount'])/df_true.shape[0]

# Error analysis
print('Accuracy')
print(amount_acc)
col = 'amount'
col_pred = 'amount_pred'
true_res = df_true[df_pred[col_pred] != df_true[col]][col]
pred_res = df_pred[df_pred[col_pred] != df_true[col]][col_pred]
df_error_analysis = pd.concat([true_res, pred_res], axis=1)
print(df_error_analysis)
```
